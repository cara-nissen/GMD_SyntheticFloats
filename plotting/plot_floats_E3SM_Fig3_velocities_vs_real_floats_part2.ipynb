```python
#----
# FUNCTIONS
#----

@njit
def get_distance_between_two_points(lon_point, lat_point, lon_point2, lat_point2):
        # all coordinates NOT in radians
        #from math import sin, cos, sqrt, atan2, radians
        #import numpy as np
        # approximate radius of earth in km
        R = 6373.0

        #lat2 = radians(mesh.y2) # all positions in mesh
        #lon2 = radians(mesh.x2)
        #lat2 = [radians(x) for x in mesh.y2]
        #lon2 = [radians(x) for x in mesh.x2]
        lat1 = radians(lat_point)
        lon1 = radians(lon_point)
        bb1 = cos(lat1)
        lat2 = radians(lat_point2)
        lon2 = radians(lon_point2)

        dlon = lon2 - lon1
        dlat = lat2 - lat1
        a = sin(dlat / 2)**2 + bb1 * cos(lat2) * sin(dlon / 2)**2
        distance = 2*R*atan2(sqrt(a), sqrt(1 - a)) # to speed things up, omit constant factors here!

        return distance
```

```python
#---
# plot map of velocities at 1000m: step2
#---

# Load data from input file (model output)
varname1 = 'timeMonthly_avg_velocityMeridional'
varname2 = 'timeMonthly_avg_velocityZonal'
print(zlevs[38])

velocity = np.zeros(area.shape[0])
for yy in range(0,len(year_list)):
    print('Load year '+year_list[yy])
    inputfile_run1 = inputroot+'monthlyAverageEulerianFields.year'+year_list[yy]+'.nc'
    finput_run1 = xr.open_dataset(inputfile_run1)

    # MPAS-O mesh for EC30to60E2r2 run 1
    vel_1 = finput_run1[varname1].data 
    vel_1 = np.squeeze(vel_1)[:,:,38].mean(axis=0) # annual mean, index 38 = 928m
    #vel_1
    vel_2 = finput_run1[varname2].data
    vel_2 = np.squeeze(vel_2)[:,:,38].mean(axis=0) # annual mean, index 38 = 928m
    #print(vel_1.shape,vel_2.shape)

    velocity = velocity+np.sqrt(np.multiply(vel_1,vel_1) + np.multiply(vel_2,vel_2))
    print(velocity.shape,np.min(velocity),np.max(velocity))

# get avg velocity field
print('6-year average:')
velocity = np.divide(velocity,len(year_list))
print(velocity.shape,np.min(velocity),np.max(velocity))
```

```python
#----
# get distance travelled along each trajectory, ignore "zeros" in output
#----
# correction July 11, 2023: don't delete zeros but set these to NaN
# by deleting zeros, I increase the distance travelled, leading to lower velocities if assuming
# every distance has been travelled in 10 days

dist_all_E3SM = np.zeros([lat_all_10day.shape[0]-1,lat_all_10day.shape[1]]) # num_time_steps minus 1 x num_particles

for nn in range(0,lat_all_10day.shape[1]): # loop over flots
    if np.mod(nn,1000)==0:
        print('Process float number',nn+1)
    lon_pp = lon_all_10day[:,nn]
    lat_pp = lat_all_10day[:,nn]
    # kick out zeros (if any)
    ind0 = np.where(lon_pp==0)[0]
    lon_pp[ind0] = np.nan # set to NaN to not compute a distance 
    lat_pp[ind0] = np.nan
    #lon_pp = np.delete(lon_pp,ind0)
    #lat_pp = np.delete(lat_pp,ind0)
    # del ind0
    for tt in range(0,lat_pp.shape[0]-1): # loop over time
        lon_point1,lat_point1 = lon_pp[tt],lat_pp[tt] #lon_all[tt,0,nn],lat_all[tt,0,nn]
        lon_point2,lat_point2 = lon_pp[tt+1],lat_pp[tt+1] #lon_all[tt+1,0,nn],lat_all[tt+1,0,nn]
        #print(lon_point2,lat_point2)
        dist_all_E3SM[tt,nn] = get_distance_between_two_points(lon_point1,lat_point1,\
                                                          lon_point2,lat_point2)
        del lon_point1,lat_point1,lon_point2,lat_point2
    del lon_pp,lat_pp

# set zeros to NaN (not interested in floats that don't move -> these are the ones that enter too shallow areas)
dist_all_E3SM[dist_all_E3SM<=0.1]=np.nan

# convert from distance in km per 10 days to velocity in cm s-1
factor_vel = 100*1000/(24*10*3600) 

print('Min/Max distance:',np.nanmin(dist_all_E3SM),np.nanmax(dist_all_E3SM))
print('Min/Max velocity in cm s-1:',np.nanmin(dist_all_E3SM*factor_vel),np.nanmax(dist_all_E3SM*factor_vel))

lat_all_10day_b = lat_all_10day[:-1,:]
lon_all_10day_b = lon_all_10day[:-1,:]

print('done')
```

```python
#---
# statistics
# functions from here: https://stackoverflow.com/questions/38641691/weighted-correlation-coefficient-with-pandas
#---
def m(x, w):
    """Weighted Mean"""
    return np.sum(x * w) / np.sum(w)

def cov(x, y, w):
    """Weighted Covariance"""
    return np.sum(w * (x - m(x, w)) * (y - m(y, w))) / np.sum(w)

def corr(x, y, w):
    """Weighted Correlation"""
    return cov(x, y, w) / np.sqrt(cov(x, x, w) * cov(y, y, w))

#----
# error statistics
# functions from here: https://gist.github.com/bshishov/5dc237f59f019b26145648e2124ca1c9
#----

def _error(actual: np.ndarray, predicted: np.ndarray):
            """ Simple error """
            return actual - predicted

def mse(actual: np.ndarray, predicted: np.ndarray):
    """ Mean Squared Error """
    return np.mean(np.square(_error(actual, predicted)))

def rmse(actual: np.ndarray, predicted: np.ndarray):
    """ Root Mean Squared Error """
    return np.sqrt(mse(actual, predicted))

def nrmse(actual: np.ndarray, predicted: np.ndarray):
    """ Normalized Root Mean Squared Error """
    #print(np.var(actual))
    return rmse(actual, predicted) / np.std(actual) #(actual.max() - actual.min())

print()
print('Float-based: E3SM vs Argo')
print()
#---
# correlation 
#---
print('Area-weighted correlation coefficient R:',corr(aux1,aux2,aux3/np.sum(aux3)))
print('R2:',corr(aux1,aux2,aux3/np.sum(aux3))*corr(aux1,aux2,aux3/np.sum(aux3)))

#---
# RMSE
#---
#diff_aux = aux1 - aux2 # E3SM minus Scripps
#print('RMSE:',np.sum(diff_aux*aux3)/np.sum(aux3))
weighted_mean_sq_error_sklearn = np.average( 
    (aux1-aux2)**2, axis=0, weights=aux3/np.sum(aux3))
print('Area-weighted RMSE:',np.sqrt(weighted_mean_sq_error_sklearn))

#---
# BIAS (Lagrangian E3SM minus Scripps)
#---
diff_aux = aux1 - aux2 # E3SM minus Scripps
print('Area-weighted mean bias:',np.sum(diff_aux*aux3)/np.sum(aux3))

print()
print('Float-based E3SM vs Eulerian E3SM')
print()
#---
# correlation 
#---
print('Area-weighted correlation coefficient R:',corr(aux1b,aux4b,aux3b/np.sum(aux3b)))
print('R2:',corr(aux1b,aux4b,aux3b/np.sum(aux3b))*corr(aux1b,aux4b,aux3b/np.sum(aux3b)))

#---
# RMSE
#---
#diff_aux = aux1 - aux2 # E3SM minus Scripps
#print('RMSE:',np.sum(diff_aux*aux3)/np.sum(aux3))
weighted_mean_sq_error_sklearn = np.average( 
    (aux1b-aux4b)**2, axis=0, weights=aux3b/np.sum(aux3b))
print('Area-weighted RMSE:',np.sqrt(weighted_mean_sq_error_sklearn))

#---
# BIAS (Lagrangian E3SM minus Scripps)
#---
diff_aux = aux1b - aux4b # E3SM minus Scripps
print('Area-weighted mean bias:',np.sum(diff_aux*aux3b)/np.sum(aux3b))
```
