```python

import sys
sys.path.append('/global/homes/c/cnissen/scripts/seawater-3.3.4/seawater/')
import os
import numpy as np
import xarray as xr
import cartopy
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
from matplotlib import cm
import seawater
#from seawater import dist
#import seawater as sw
import matplotlib.path as mpath
from cartopy.util import add_cyclic_point
import matplotlib.gridspec as gridspec
import matplotlib.ticker as mticker
from cartopy.mpl.ticker import (LongitudeFormatter, LatitudeFormatter,
                                LatitudeLocator)
import random
from numba import njit
from math import sin, cos, sqrt, atan2, radians
from argopy import DataFetcher as ArgoDataFetcher
```

```python

argo_loader = ArgoDataFetcher()
#ddd = argo_loader.region([-180, 180, -90, 90, 0, 10, '2012-01-01', '2013-01-01']) 
ddd = argo_loader.region([-180, 180, 0, 10, 0, 10, '2012-01', '2013-01'])

print(ddd)

ds_points = ddd.to_xarray()
ds_profiles = ds_points.argo.point2profile()
print(ds_profiles)
```

```python

cycles = ds_profiles['CYCLE_NUMBER'].values
lat    = ds_profiles['LATITUDE'].values
lon    = ds_profiles['LONGITUDE'].values
pres   = ds_profiles['PRES'].values
print(cycles.shape,lat.shape,lon.shape,pres.shape)

first_cycle = np.where(cycles==0)[0]

print('Number of first cycles: ',first_cycle.shape)
print(lat[first_cycle],lon[first_cycle])

```

```python
#----
# loop over all lat/lon and years
#----

# initialize arrays
lon_float_all  = []
lat_float_all  = []

year_start, year_end = 2021, 2022

res = 2.
#lat_start_list = [-75,-73,-71] #np.arange(-80.,88.+res,res)
#lat_end_list   = [-73,-71,-69] #np.arange(-78.,90.+res,res)

lat_start_list = np.arange(-90.,90-res+res,res)
lat_end_list   = np.arange(-90+res,90.+res,res)

argo_loader = ArgoDataFetcher()

for mm in range(0,len(lat_start_list)):
    print('Process lat_start '+str(lat_start_list[mm])+', lat_end '+str(lat_end_list[mm]))
    
    try:
        ddd = argo_loader.region([-180, 180, lat_start_list[mm], lat_end_list[mm], 0, 10,\
                                  str(year_start)+'-01', str(year_end)+'-01'])
        #print(ddd)

        #print('Convert to xarray...')
        ds_profiles = ddd.to_xarray()
        #ds_profiles = ds_profiles.argo.point2profile()
        #print(ds_profiles)

        cycles = ds_profiles['CYCLE_NUMBER'].values
        lat    = ds_profiles['LATITUDE'].values
        lon    = ds_profiles['LONGITUDE'].values

        first_cycle = np.where(cycles==1)[0]
        #print(lat[first_cycle],lon[first_cycle])

        if first_cycle.shape[0]>0:
            lat_float_all  = np.hstack((lat_float_all,lat[first_cycle]))
            lon_float_all  = np.hstack((lon_float_all,lon[first_cycle]))
    except:
        print('No data found')
        pass
    
print(len(lat_float_all))

#---
# get unique positions (by choosing depth levels 0-10, some floats have multiple entries for the first complete cycle)
#---

positions = np.vstack((lat_float_all,lon_float_all)).transpose()
print(positions.shape)

unique_rows = np.unique(positions, axis=0)
print(unique_rows.shape)

#---
# save as txt file
#---
print('Save unique lat/lon positions of first complete cycle in txt files...')

savepath1 = '/global/cscratch1/sd/cnissen/ARGO_floats/txt_files/'

with open(savepath1+'Argo_floats_latitude_first_cycle_year_'+str(year_start)+'.txt',"w") as output:
    for line in unique_rows[:,0]:
        output.write(f"{line}\n")
    #output.write(lat_float_all)
    
with open(savepath1+'Argo_floats_longitude_first_cycle_year_'+str(year_start)+'.txt',"w") as output:
    for line in unique_rows[:,1]:
        output.write(f"{line}\n")
    
print('done')


```
